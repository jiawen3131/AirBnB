## Regression Model

```r
london_weekday <- read.csv('london_weekday_cleaned_final.csv')
str(london_weekday)
```

```r
london_weekdaylogprice <- log(london_weekday$realSum)
london_weekdaylogMetroDist <- log(london_weekday$metro_dist)
london_weekdaylogGuestSatisfaction <- log(london_weekday$guest_satisfaction_overall)
london_weekdaylogcleanliness_rating <- london_weekday$cleanliness_rating
london_weekdaydist<- london_weekday$dist
london_weekdayattr_index_norm <- london_weekday$attr_index_norm
london_weekdayrest_index_norm <- london_weekday$rest_index_norm 
```

### Log(MetroDist)
Y =  5.117987 - 0.114686X = 5.12 - 0.115X
<br>
p-value = <2.2e-16
<br>
R-squared = 0.1161

```r
library(repr)
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)
MetroDistModel <- lm(log(london_weekday$realSum)~log(london_weekday$metro_dist))
summary(MetroDistModel)

#Correlation
cor(london_weekdaylogprice,london_weekdaylogMetroDist)

#Regression line
plot(london_weekdaylogMetroDist,london_weekdaylogprice, col= "blue" )
abline(lm(london_weekdaylogprice ~london_weekdaylogMetroDist))

#qqplot of residuals
qqnorm(MetroDistModel$residuals)
qqline(MetroDistModel$residuals)

```

### Dist (aka Distance to city)
Y = 5.45 - 0.0504X
<br>
p-value = <2.2e-16
<br>
R-squared = 0.182

```r
library(repr)
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)
DistModel <- lm(log(london_weekday$realSum)~london_weekday$dist)
summary(DistModel)

#Correlation
cor(london_weekdaylogprice,london_weekdaydist)

#Regression line
plot(london_weekdaydist,london_weekdaylogprice, col= "blue" )
abline(lm(london_weekdaylogprice ~london_weekdaydist))

#qqplot of residuals
qqnorm(DistModel$residuals)
qqline(DistModel$residuals)
```

### log(Attr_index_norm)

<br>
R-squared = 0.1649

```r
library(repr)
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)
AttrModel <- lm(log(london_weekday$realSum)~log(london_weekday$attr_index_norm))
summary(AttrModel)

#Correlation
cor(london_weekdaylogprice,log(london_weekdayattr_index_norm))

#Regression
plot(log(london_weekdayattr_index_norm), london_weekdaylogprice, col= "blue" )
abline(lm(london_weekdaylogprice ~ log(london_weekdayattr_index_norm)))

#qqplot of residuals
qqnorm(AttrModel$residuals)
qqline(AttrModel$residuals)
```

### log(Rest_index_norm)
Y = 4.491 + 0.307X
<br>
p-value = <2.2e-16
<br>
R-squared = 0.1626

```r
library(repr)
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)
RestModel <- lm(log(london_weekday$realSum)~log(london_weekday$rest_index_norm))
summary(RestModel)

#Correlation
cor(london_weekdaylogprice,log(london_weekday$rest_index_norm))

#Regression
plot(log(london_weekdayrest_index_norm), london_weekdaylogprice,col= "cornflowerblue" )
abline(lm(london_weekdaylogprice ~ log(london_weekday$rest_index_norm)))

#qqplot of residual
qqnorm(RestModel$residuals)
qqline(RestModel$residuals)
```

### Multiple Linear Regression

```r
finalModel <- lm(log(london_weekday$realSum)~ log(london_weekday$rest_index_norm) + log(london_weekday$attr_index_norm) +
london_weekday$dist + log(london_weekday$metro_dist))
step(finalModel, direction = 'backward')
```

```r
newx <- data.frame(logrest_index_norm=log(40), logattr_index_norm=log(60), dist=0.2 , logmetro_dist=log(0.8))
price <- log(london_weekday$realSum)
logrest_index_norm <- log(london_weekday$rest_index_norm)
logattr_index_norm <- log(london_weekday$attr_index_norm)
logmetro_dist <- log(london_weekday$metro_dist)
dist <- log(london_weekday$dist)
predm <- predict(lm(price~logrest_index_norm+logattr_index_norm+dist+logmetro_dist), newx, level =0.95, interval = 'confidence')
cbind(newx,round(exp(predm),2))
```
